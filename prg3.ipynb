```python
#3. Write a program to create interactive visualizations using Plotly, including interactive
# line, bar, and scatter plots with hover information.

import pandas as pd
import plotly.express as px

# Load dataset
df = pd.read_csv("gapminder.csv")

# 1. Line Chart: Life Expectancy Over Time (Oceania)
df_oceania = df.query("continent == 'Oceania'")

fig_line = px.line(
    df_oceania,
    x="year",
    y="lifeExp",
    color="country",
    markers=True,
    title="Life Expectancy Over Time (Oceania)",
    labels={
        "lifeExp": "Life Expectancy",
        "year": "Year"
    },
    hover_name="country"
)

# 2. Bar Chart: GDP per Capita in Europe (2007)
df_2007_europe = df.query("year == 2007 and continent == 'Europe'")

fig_bar = px.bar(
    df_2007_europe,
    x="country",
    y="gdpPercap",
    color="gdpPercap",
    title="GDP per Capita in Europe (2007)",
    labels={"gdpPercap": "GDP per Capita"},
    hover_data=["pop"]
)

# 3. Scatter Plot: Global Health vs Wealth (2007)
df_2007 = df.query("year == 2007")

fig_scatter = px.scatter(
    df_2007,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    hover_name="country",
    log_x=True,
    size_max=60,
    title="Global Health vs Wealth (2007)",
    labels={
        "gdpPercap": "GDP per Capita (Log Scale)",
        "lifeExp": "Life Expectancy"
    }
)

# Show all plots
fig_line.show()
fig_bar.show()
fig_scatter.show()
```

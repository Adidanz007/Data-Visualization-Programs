```python
# 5. Write a program to visualize geographical data, using map-based plots to represent location-based information.
import pandas as pd
import plotly.express as px

# Step 1: Load the dataset
url = "earthquakes-23k.csv"
df = pd.read_csv(url)

# Step 2: Display dataset information
print("Dataset Preview:")
print(df.head())

print("\nDataset Info:")
df.info()

# Step 3: Clean dataset (remove missing values)
df = df.dropna(subset=["Latitude", "Longitude", "Magnitude"])

# Step 4: Create a map visualization
fig = px.scatter_geo(
    df,
    lat="Latitude",
    lon="Longitude",
    size="Magnitude",
    color="Magnitude",
    hover_name="Magnitude",
    projection="natural earth",
    title="Global Earthquake Distribution Map"
)

# Step 5: Show the map
fig.show()
```

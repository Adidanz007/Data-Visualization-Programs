```python
# 4. Write a program to design an interactive dashboard using Plotly, incorporating multiple charts, filters, and dropdown controls.  (Dataset : tips.csv)
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html, Input, Output

# Load data
df = pd.read_csv("tips.csv")

# App
app = Dash(__name__)

app.layout = html.Div([
    html.H2("Restaurant Tips Dashboard", style={"textAlign": "center"}),

    dcc.Dropdown(
        id="day",
        options=[{"label": d, "value": d} for d in df["day"].unique()],
        value=df["day"].unique()[0],
        clearable=False
    ),

    dcc.Dropdown(
        id="time",
        options=[{"label": t, "value": t} for t in df["time"].unique()],
        value=df["time"].unique()[0],
        clearable=False
    ),

    dcc.Graph(id="scatter"),
    dcc.Graph(id="bar"),
    dcc.Graph(id="hist"),
    dcc.Graph(id="box")
])

@app.callback(
    [Output("scatter", "figure"),
     Output("bar", "figure"),
     Output("hist", "figure"),
     Output("box", "figure")],
    [Input("day", "value"),
     Input("time", "value")]
)
def update(day, time):
    dff = df[(df["day"] == day) & (df["time"] == time)]

    return (
        px.scatter(dff, x="total_bill", y="tip", color="sex", size="size"),
        px.bar(dff, x="sex", y="tip", color="smoker"),
        px.histogram(dff, x="total_bill", color="sex"),
        px.box(dff, x="sex", y="tip", color="sex")
    )

if __name__ == "__main__":
    app.run(debug=True)
```

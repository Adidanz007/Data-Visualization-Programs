```python
# 2. Write a program to create pair plots and regression plots using Seaborn, to study variate relationships and trends. (for all 3 Datasets)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set styling
sns.set_theme(style="whitegrid")

# 1. Load Dataset & Feature Engineering
iris = pd.read_csv("iris.csv")
print(iris.head())
iris["sepal_ratio"] = iris["sepal_length"] / iris["sepal_width"]
iris["petal_ratio"] = iris["petal_length"] / iris["petal_width"]

# 2. Advanced Correlation Heatmap
plt.figure(figsize=(8, 6))
corr = iris.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Advanced Correlation Heatmap")
plt.show()

# 3. Custom PairGrid (Advanced Pair Plot)
g = sns.PairGrid(iris, hue="species", height=2.5)
g.map_diag(sns.histplot, kde=True)
g.map_upper(sns.scatterplot)
g.map_lower(
    sns.regplot,
    order=2,
    scatter_kws={"s": 20, "alpha": 0.4}
)
g.add_legend()
plt.show()

# 4. Polynomial Regression (Degree 3)
plt.figure(figsize=(7, 5))
sns.regplot(
    x="sepal_length",
    y="petal_length",
    data=iris,
    order=3,
    scatter_kws={"s": 40},
    line_kws={"color": "red"},
    ci=95
)
plt.title("Polynomial Regression (Degree 3)")
plt.show()

# 5. Statistical Analysis
corr_val, p_val = stats.pearsonr(
    iris["sepal_length"],
    iris["petal_length"]
)
print(f"Pearson Correlation: {round(corr_val, 4)}")
print(f"P-value (Significance): {round(p_val, 6)}")
```
